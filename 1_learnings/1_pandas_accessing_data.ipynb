```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt


# Loading Dataset
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

### .iloc[]
interger based , row/column (df.iloc[row, column]). , end is exclusive

### .loc[]
label based , row/column (df.loc[row, column]). , end is **inclusive**

```python
df.loc[df['job_title_short'] == 'Data Analyst']
```

```python
# using .loc instead of .iloc for accessing the data easily

df.loc[0:, 'salary_rate':'salary_hour_avg'].dropna(subset='salary_rate')
```

```python
df.loc[0:5,['job_title_short','job_title','salary_year_avg']]
```
