```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt


# Loading Dataset
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

NOTE: Use pie charts for comparing two variables

don't use it for more than two variables because its hard to read instead use bar. 

```python
df['job_work_from_home'].value_counts().plot(kind='pie', startangle=90, autopct='%1.1f%%')
plt.title('Job Work from Home Status')
plt.ylabel('')
plt.show()

# NOTE: autopct='%1.1f%%' = here (1.1 = decimal places) ,  (f = float), (%% = to add %)
```

```python
fig, ax = plt.subplots(1,3)

dict_column = {
    'job_work_from_home' : 'Work from Home',
    'job_no_degree_mention' : 'Job Degree Req.',
    'job_health_insurance' : 'Heath Insurance offered'
}

for i, (column,title) in enumerate(dict_column.items()):
    ax[i].pie(df[column].value_counts(), startangle=90, autopct='%1.1f%%', labels=['False', 'True'])
    ax[i].set_title(title)

# NOTE: here labels=['False', 'True'] if you reverse them so it reverse the true for false and 
# vice versa provide bad data 
```
